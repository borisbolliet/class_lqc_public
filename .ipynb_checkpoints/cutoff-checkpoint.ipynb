    s = 0.
    j = 0
    while j<bin_size:
        s += Ps[i+j]
        j+=1
    j = 1
    while j<bin_size:
        s += Ps[i-j]
        j+=1
    Ps_mean.append(s/(2.*bin_size+1.))
    i+=1    
Ps_mean = np.asarray(Ps_mean)
