```python
#Reading data from files and plotting them
#read from textfile to x and y
#/phi_b = 1.19
data_sps = genfromtxt('job30315/output/test-_pk_lqc_s.dat')
kp1 = 2.353517103813e+03 
k_1 = data_sps[:,0]
Ps_1 = data_sps[:,1]
data_sps = genfromtxt('job30316/output/test-_pk_lqc_s.dat')
kp1 = 2.353517103813e+03 
k_1a = data_sps[:,0]
Ps_1a = data_sps[:,1]
```

```python
#Reading data from files and plotting them
#read from textfile to x and y
#/phi_b = 1.17
data_sps = genfromtxt('job30308/output/test-_pk_lqc_s.dat')
kp2 = 1.078717683331e+03 
k_2 = data_sps[:,0]
Ps_2 = data_sps[:,1]
data_sps = genfromtxt('job30317/output/test-_pk_lqc_s.dat')
kp2 = 1.078717683331e+03 
k_2a = data_sps[:,0]
Ps_2a = data_sps[:,1]
```

```python
#Reading data from files and plotting them
#read from textfile to x and y
#/phi_b = 1.15
data_sps = genfromtxt('job30309/output/test-_pk_lqc_s.dat')
kp3 =  4.749130224196e+02
k_3 = data_sps[:,0]
Ps_3 = data_sps[:,1]
data_sps = genfromtxt('job30318/output/test-_pk_lqc_s.dat')
kp3 =  4.749130224196e+02
k_3a = data_sps[:,0]
Ps_3a = data_sps[:,1]
```

```python
#Reading data from files and plotting them
#read from textfile to x and y
#/phi_b = 1.13
data_sps = genfromtxt('job30310/output/test-_pk_lqc_s.dat')
kp4 =   2.080259278004e+02
k_4 = data_sps[:,0]
Ps_4 = data_sps[:,1]
data_sps = genfromtxt('job30319/output/test-_pk_lqc_s.dat')
kp4 =   2.080259278004e+02
k_4a = data_sps[:,0]
Ps_4a = data_sps[:,1]
```

```python
#plot(t, a, color='red', linestyle='solid',label=r'$a$')
#plot(t, H, color='red', linestyle='--',label=r'$H$')
#plot(t, rho, color='blue', linestyle='solid',label=r'$\rho$')
plot(k_1/kp1, Ps_1, color='k', linestyle='None',marker='o',markeredgewidth=0.,markersize = 4,label=r'$\phi_b = 1.19 m_{_{\rm Pl}}$')
plot(k_1a/kp1, Ps_1a, color='k', linestyle='None',marker='o',markeredgewidth=0.,markersize = 4)
plot(k_2/kp2, Ps_2, color='slategrey', linestyle='None',marker='v',markeredgewidth=0.,markersize = 4,label=r'$\phi_b = 1.17 m_{_{\rm Pl}}$')
plot(k_2a/kp2, Ps_2a, color='slategrey', linestyle='None',marker='v',markeredgewidth=0.,markersize = 4)
plot(k_3/kp3, Ps_3, color='grey', linestyle='None',marker='+',markeredgewidth=1.,markersize = 4,label=r'$\phi_b = 1.15 m_{_{\rm Pl}}$')
plot(k_3a/kp3, Ps_3a, color='grey', linestyle='None',marker='+',markeredgewidth=1.,markersize = 4)
plot(k_4/kp4, Ps_4, color='lightgrey', linestyle='None',marker='*',markeredgewidth=0.,markersize = 4,label=r'$\phi_b = 1.13 m_{_{\rm Pl}}$')
plot(k_4a/kp4, Ps_4a, color='lightgrey', linestyle='None',marker='*',markeredgewidth=0.,markersize = 4)
pylab.annotate(r'$\rho_b = 0.41 m_{_{\rm Pl}}^4$', xy=(1.e-1,2.e-8),  xycoords='data',fontsize=12)
#plot(k_p1/kp, Ps_p1, color='grey', linestyle='None',marker='o',markeredgewidth=0.,markersize = 4)
#plot(k_o/kp, Ps_o, color='red', linestyle='-',linewidth=0.25)
#plot([1.607738e+01/kp,1.607738e+01/kp],[1e-12,1e-6],linestyle='--',color='black',linewidth=0.5)
#pylab.annotate(r'$k_{_{\rm LQC}}$', xy=(275,290),  xycoords='figure points',fontsize=12)
plt.xscale('log')
plt.yscale('log')
legend=pylab.legend(loc='best', numpoints=1, shadow=False,frameon=True,prop={'size':13})
pylab.xlabel(r'$k/k_*$')
pylab.ylabel(r'${\cal P}_s(k)$')
pylab.xlim([1.e-7,1.e1])
pylab.ylim([1e-12,1e-7])
pylab.savefig('Ps-phib.pdf', bbox_inches='tight')
#pylab.show()
```
