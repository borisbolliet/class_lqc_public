```python
from scipy.interpolate import InterpolatedUnivariateSpline
from scipy.interpolate import interp1d
bin_size = 20
Ps_mean = []
i = bin_size
while i+bin_size<len(Ps):
    s = 0.
    j = 0
    while j<bin_size:
        s += Ps[i+j]
        j+=1
    j = 1
    while j<bin_size:
        s += Ps[i-j]
        j+=1
    Ps_mean.append(s/(2.*bin_size+1.))
    i+=1    
Ps_mean = np.asarray(Ps_mean)

k_mean = k[bin_size:len(k)-bin_size]
f3= interp1d(k_mean, Ps_mean)

k_new = np.logspace(np.log10(k[bin_size + 1]),np.log10(k[len(k_mean)-1]),1000)
Ps_new = f3(k_new)
```
