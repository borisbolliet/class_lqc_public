```python
#plot(t, a, color='red', linestyle='solid',label=r'$a$')
#plot(t, H, color='red', linestyle='--',label=r'$H$')
#plot(t, rho, color='blue', linestyle='solid',label=r'$\rho$')
plot(k/kp, P, color='red', linestyle='-',label=r'$new$')
plot(k_b/kp, P_b, color='blue', linestyle='--',label=r'$old$')
plt.xscale('log')
plt.yscale('log')
#plt.xscale('symlog')
#plot(t, phi_dot, color='blue', linestyle='-.',label=r'$\dot\phi$')
pylab.xlabel(r'$k/k_{\star}$')
pylab.ylabel(r'${\cal P}_s(k)$')
legend=pylab.legend(loc='best', shadow=False,frameon=False,prop={'size':13})
#pylab.xlim([-1.e7,1.e7])
#pylab.ylim([-0.5,4])
#pylab.savefig('fnl-lqc-sl-slowroll-x_ini_3p19em5.pdf', bbox_inches='tight')
#pylab.show()
```
