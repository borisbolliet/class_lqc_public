```python
PATH = 'output/test-fb-d_10-perturbations_lqc_k0_s.dat'
T0a = loadtxt(PATH)

PATH = 'output/test-fb-d_100-perturbations_lqc_k0_s.dat'
T0b = loadtxt(PATH)

PATH = 'output/test-fb-d_1000-perturbations_lqc_k0_s.dat'
T0c = loadtxt(PATH)
#
PATH = 'output/test-mb-d_10-perturbations_lqc_k0_s.dat'
T1a = loadtxt(PATH)

PATH = 'output/test-mb-d_100-perturbations_lqc_k0_s.dat'
T1b = loadtxt(PATH)

PATH = 'output/test-mb-d_1000-perturbations_lqc_k0_s.dat'
T1c = loadtxt(PATH)
```

```python
k = [20.,20.,20.,20.,20.,20.]
```

```python
#TT=[T0,Tdef,Tm1000,Tm100,Tm10,T5,T10]

TT=[T0a,T0b,T0c,T1a,T1b,T1c]

cosmic_time = []
efolds =[]
fNL = []
pre123 = []
pre1d23 = []
pre1d2d3 = []
Qre = []
Qim = []
Qpre = []
Qpim = []

for i in range (len(TT)):
    T = TT[i]
    a_over_z = T[:,47]
    
    Y_000 = T[:,24]*a_over_z*a_over_z*a_over_z
    Y_100 = T[:,25]*a_over_z*a_over_z*a_over_z
    Y_110 = T[:,26]*a_over_z*a_over_z*a_over_z
    Y_111 = T[:,27]*a_over_z*a_over_z*a_over_z
    
    Y_011 = Y_110
    Y_101 = Y_110
    
    Y_010 = Y_100
    Y_001 = Y_100


    Z_123_000 =  T[:,12]
    Z_123_100 =  T[:,13]
    Z_123_110 =  T[:,14]
    Z_123_111 =  T[:,15]

    Z_123_010 = Z_123_100
    Z_123_001 = Z_123_100

    Z_123_011 = Z_123_110
    Z_123_101 = Z_123_110
    
    
    
    Z_123d_000 =  T[:,16]
    Z_123d_100 =  T[:,17]
    Z_123d_110 =  T[:,18]
    Z_123d_111 =  T[:,19]
    
    
    Z_123d_010 = Z_123d_100
    Z_123d_001 = T[:,35]
    
    Z_123d_011 = T[:,36]
    Z_123d_101 = Z_123d_011
    
    
    
    
    
    Z_12d3d_000 =  T[:,20]
    Z_12d3d_100 =  T[:,21]
    Z_12d3d_110 =  T[:,22]
    Z_12d3d_111 =  T[:,23]
    
    Z_12d3d_010 = T[:,39]
    Z_12d3d_001 = Z_12d3d_010
    
    Z_12d3d_011 = T[:,44]
    Z_12d3d_101 = Z_12d3d_110

    
    
    
    
    
    
    integral_123 =\
          2.*(Y_000-Y_011-Y_101-Y_110)\
          *(\
            Z_123_111\
            -Z_123_001\
            -Z_123_010\
            -Z_123_100\
          )\
          -2.*(Y_111-Y_100-Y_010-Y_001)\
          *(Z_123_000\
            -Z_123_110\
            -Z_123_101\
           -Z_123_011)  
    
    integral_123d =\
          2.*(Y_000-Y_011-Y_101-Y_110)\
          *(Z_123d_111\
            -Z_123d_001\
            -Z_123d_010\
            -Z_123d_100)\
          -2.*(Y_111-Y_100-Y_010-Y_001)\
          *(Z_123d_000\
            -Z_123d_110\
            -Z_123d_101\
            -Z_123d_011)
            
    integral_12d3 = integral_123d
    integral_1d23 = integral_123d
    
    integral_12d3d =\
          2.*(Y_000-Y_011-Y_101-Y_110)\
          *(\
            Z_12d3d_111\
          -Z_12d3d_001\
          -Z_12d3d_010\
          -Z_12d3d_100)\
          -2.*(Y_111-Y_100-Y_010-Y_001)\
           *(Z_12d3d_000\
           -Z_12d3d_110\
           -Z_12d3d_101\
          -Z_12d3d_011)
    
    integral_1d23d = integral_12d3d
    integral_1d2d3 = integral_12d3d
    
        
    field_redef = T[:,28]
    
          
    
    integral =(integral_123 \
          +integral_1d23\
          +integral_12d3\
          +integral_123d \
          +integral_1d2d3\
          +integral_1d23d\
          +integral_12d3d\
          +field_redef)
    
                        
                
    pk_s_1 = T[:,6]
    pk_s_2 = pk_s_1
    pk_s_3 = pk_s_1
    
    k_1 = k[i]
    k_2 = k_1
    k_3 = k_1
    
    
    N = T[:,29]
    integral_s= -10./3.\
  *pow(2.*np.pi,-4.)\
  *pow(k_1*k_2*k_3,3.)\
  *integral\
  *pow(pow(k_1,3.)*pk_s_2*pk_s_3\
       +pow(k_2,3.)*pk_s_1*pk_s_3\
       +pow(k_3,3.)*pk_s_2*pk_s_1,-1.)
    
    efolds.append(N)
    fNL.append(integral_s)

    P_123 = T[:,48]
    P_1d23 = T[:,49]
    P_1d2d3 = T[:,50]

    pre123.append(P_123)
    pre1d23.append(P_1d23)
    pre1d2d3.append(P_1d2d3)
    
    time = T[:,0]
    cosmic_time.append(time)
    Qre.append(T[:,2])
    Qim.append(T[:,3])
    Qpre.append(T[:,4])
    Qpim.append(T[:,5])
```
