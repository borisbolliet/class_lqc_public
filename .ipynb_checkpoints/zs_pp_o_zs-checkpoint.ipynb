```python
#Reading data from files and plotting them
#read from textfile to x and y
data_bg = genfromtxt('output/test-phasespace-positive_branch-040617-_background_lqc.dat')
t = data_bg[:,0]
a = data_bg[:,1]
x = data_bg[:,2]
y = data_bg[:,3]
H = data_bg[:,4]
epsilon_H = data_bg[:,5]
epsilon_2 = data_bg[:,6]
rho = data_bg[:,7]
phi = data_bg[:,9]
phi_dot = data_bg[:,10]
#f_S = data_bg[:,13]
H2_f_S_Q = data_bg[:,11]
N = data_bg[:,14]
zS_primeprime_over_zS = data_bg[:,17]
pia_phase = data_bg[:,23]
pia_phys = data_bg[:,24]
H2_f_S_Q_phys = data_bg[:,25]
V= data_bg[:,26]
cosmic_time = data_bg[:,27]
kp=1.705662e+02**2
kp_list = np.array([kp]*len(N))
```
