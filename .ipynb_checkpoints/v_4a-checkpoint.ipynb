```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.interpolate import spline
import numpy as np
from numpy import *
import pylab
import scipy
from matplotlib import rc, rcParams, pyplot
from matplotlib.pyplot import figure, axes, plot, xlabel, ylabel, title, grid, savefig, show
from matplotlib.ticker import MaxNLocator
######################################################################
figure(1,figsize=(8,6))
######################################################################
rc('font', family='TimesRoman', weight = 'extra bold', size = 18.0)
rc('text', usetex=True)
rc('axes', linewidth = 1, labelsize = 'large')  
rc('xtick', labelsize= 'medium')
rcParams['xtick.major.size'] = 6. 
rcParams['xtick.minor.size'] = 3
rcParams['xtick.major.pad'] = 12 
rcParams['xtick.minor.pad'] = 8
rc('ytick', labelsize= 'medium')  
rcParams['ytick.major.size'] = 6 
rcParams['ytick.minor.size'] = 3
rcParams['ytick.major.pad'] = 12 
rcParams['ytick.minor.pad'] = 8
rc('lines', linewidth = 1.5, markeredgewidth=2.0)
rc('savefig', dpi=300)
######################################################################
# Fixing space around the plot
pylab.axes([0.145,0.125,0.825,0.825])

#ax.xaxis.set_major_locator(xmajorLocator)
#ax.xaxis.set_minor_locator(xminorLocator)
#ax.yaxis.set_major_locator(ymajorLocator)
#ax.yaxis.set_minor_locator(yminorLocator)
```

```python
data = genfromtxt('output/test-fb-d_50-_pk_lqc_s.dat')
k0 = data[:,0]
Ps0 = data[:,1]
data = genfromtxt('output/test-fb-d_50-_lqc_derived_parameters.dat')
kp0 = data[0]
data = genfromtxt('output/test-fb-d_100-_pk_lqc_s.dat')
k1 = data[:,0]
Ps1 = data[:,1]
data = genfromtxt('output/test-fb-d_1000-_pk_lqc_s.dat')
k2 = data[:,0]
Ps2 = data[:,1]

plt.plot(k0/kp0,Ps0,linestyle='-',label=r'$v\_4a$')
plt.plot(k1/kp0,Ps1,linestyle='--',label=r'$v\_4a$')
plt.plot(k2/kp0,Ps2,linestyle=':',label=r'$v\_4a$')


plt.yscale('log')
plt.xscale('log')
#plt.ylim(1e-3,3e-2)
#plt.xlim(-.1,.1)
legend=plt.legend(loc='best',numpoints = 1,ncol =1, shadow=False,frameon=False,prop={'size':12})
plt.xlabel(r'$k/k_*$')
plt.ylabel(r'${\cal P}_s(k)$')

#plt.savefig('Ps.pdf', bbox_inches='tight')
#plt.show()
```

```python
data = genfromtxt('output/test-fb-d_50-_bispectrum_equi_lqc_s.dat')
k0 = data[:,0]
fnl0 = data[:,1]
data = genfromtxt('output/test-fb-d_100-_bispectrum_equi_lqc_s.dat')
k1 = data[:,0]
fnl1 = data[:,1]
data = genfromtxt('output/test-fb-d_1000-_bispectrum_equi_lqc_s.dat')
k2 = data[:,0]
fnl2 = data[:,1]

plt.plot(k0/kp0,abs(fnl0),linestyle='-',label=r'$v\_4a$')
plt.plot(k1/kp0,abs(fnl1),linestyle='--',label=r'$v\_4a$')
plt.plot(k2/kp0,abs(fnl2),linestyle=':',label=r'$v\_4a$')

plt.yscale('log')
plt.xscale('log')
plt.ylim(1e-3,1e5)
plt.xlim(3.e-4,9.e0)
legend=plt.legend(loc='best',numpoints = 1,ncol =1, shadow=False,frameon=False,prop={'size':12})
plt.xlabel(r'$k/k_*$')
plt.ylabel(r'$f_{_{\rm NL}}(k)$')

#plt.savefig('Ps.pdf', bbox_inches='tight')
#plt.show()
```
