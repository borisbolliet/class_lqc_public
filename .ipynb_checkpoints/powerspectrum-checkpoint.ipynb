```python
#Reading data from files and plotting them
#read from textfile to x and y
data_sps = genfromtxt('output/test-U2-070617-phiini_5p695-_pk_lqc_s.dat')
kp=1.858183267247e+02
k = data_sps[:,0]
Ps = data_sps[:,1]
```

```python
#plot(t, a, color='red', linestyle='solid',label=r'$a$')
#plot(t, H, color='red', linestyle='--',label=r'$H$')
#plot(t, rho, color='blue', linestyle='solid',label=r'$\rho$')
plot(k/kp, Ps, color='grey', marker='o',markeredgewidth=0.,markersize = 1)
plot([1,1],[1e-12,1e-6],linestyle='--',color='black')
plt.xscale('log')
plt.yscale('log')
legend=pylab.legend(loc='best', shadow=False,frameon=False,prop={'size':13})
pylab.xlabel(r'$k/k_*$')
pylab.ylabel(r'${\cal P}_s(k)$')
pylab.xlim([1.e-5,1.e1])
pylab.ylim([1e-10,4e-7])
#pylab.savefig('zs_pp_o_zs.pdf', bbox_inches='tight')
#pylab.show()
```

```python
from scipy.interpolate import InterpolatedUnivariateSpline
from scipy.interpolate import interp1d
bin_size = 50
Ps_mean = []
i = bin_size
while i+bin_size<len(Ps):
    s = 0.
    j = 0
    while j<bin_size:
        s += Ps[i+j]
        j+=1
    j = 1
    while j<bin_size:
        s += Ps[i-j]
        j+=1
    Ps_mean.append(s/(2.*bin_size+1.))
    i+=1    
Ps_mean = np.asarray(Ps_mean)

k_mean = k[bin_size:len(k)-bin_size]
f3= interp1d(k_mean, Ps_mean)

k_new = np.logspace(np.log10(k[bin_size + 1]),np.log10(k[len(k_mean)-1]),1000)
Ps_new = f3(k_new)
```

```python
#plot(t, a, color='red', linestyle='solid',label=r'$a$')
#plot(t, H, color='red', linestyle='--',label=r'$H$')
#plot(t, rho, color='blue', linestyle='solid',label=r'$\rho$')
plot(k/kp, Ps, color='grey', marker='o',markeredgewidth=0.,markersize = 2)
plot(k_new/kp, Ps_new, color='black',linestyle='--')
plot([1.607738e+01/kp,1.607738e+01/kp],[1e-12,1e-6],linestyle='--',color='black',linewidth=0.5)
pylab.annotate(r'$k_{_{\rm LQC}}$', xy=(1.807738e+01/kp, 2.e-10),  xycoords='data',fontsize=12)
plt.xscale('log')
plt.yscale('log')
legend=pylab.legend(loc='best', shadow=False,frameon=False,prop={'size':13})
pylab.xlabel(r'$k/k_*$')
pylab.ylabel(r'${\cal P}_s(k)$')
#pylab.xlim([5.e-4,1.e1])
pylab.ylim([1e-10,4e-7])
pylab.savefig('Ps.eps', bbox_inches='tight')
#pylab.show()
```
